```python

import pandas as pd
import selenium
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.common.exceptions import ElementNotInteractableException  
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
driver=webdriver.Chrome(r"C:\Users\YaTiKa\Desktop\chromedriver.exe")
```

```python
driver.get('https://www.ebmnews.com/2020/page/998/')
```

```python
#SCRAPING HEADLINES
start_page=998
end_page=1046
headlin=[]
for page in range(start_page,end_page+1): 
    
    
    
    headline_tags=driver.find_elements_by_xpath("//h2[@class='title']")
    for i in headline_tags:
        
        if i.text is None :
            
            headlin.append("--") 
        else:
            headlin.append(i.text)
            
                
            
            
            
        
    
    
    #for scrapping the datas from next pages.
    
    element = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "//a[@rel='next']/i")))
    driver.execute_script("arguments[0].click();", element)
    time.sleep(5)
```

```python
#SCRAPING DATES
start_page=998
end_page=1046
dt=[]
for page in range(start_page,end_page+1): 
    
    
    
    dates_tags=driver.find_elements_by_xpath("//time[@class='post-published updated']")
    for i in dates_tags:
        
        if i.text is None :
            
            
            dt.append("--") 
        else:
            dt.append(i.text)
            
                
            
            
            
        
    
    
    #for scrapping the datas from next pages.
    
    element = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "//a[@rel='next']/i")))
    driver.execute_script("arguments[0].click();", element)
    
 
```

```python
#SCRAPING AUTHORS:
start_page=998
end_page=1046
a=[]
for page in range(start_page,end_page+1): 
    
    
    
    authors_tags=driver.find_elements_by_xpath("//i[@class='post-author author']")
    for i in authors_tags:
        
        if i.text is None :
            
            
            a.append("--") 
        else:
            a.append(i.text)
            
                
            
            
            
        
    
    
    #for scrapping the datas from next pages.
    
    element = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "//a[@rel='next']/i")))
    driver.execute_script("arguments[0].click();", element)
    
 
```

```python
#SCRAPING VERTICALS
```

```python
start_page=998
end_page=1046
verticals=[]

for page in range(start_page,end_page+1): 
    
    

    urls=[i.get_attribute("href")for i in driver.find_elements_by_xpath("//a[@class='post-url post-title']")]

    
    
    for url in urls[0:10]:
        
    
        
        try:
            
            
        
            driver.get(url)
            desc=driver.find_element_by_xpath("//div[@class='term-badges ']").text
            verticals.append(desc)
        except NoSuchElementException :
            
            
        
        
            verticals.append("---")
    
        
    element = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "//a[@rel='next']/i")))
    driver.execute_script("arguments[0].click();", element)
    
  
            
            
   
```
